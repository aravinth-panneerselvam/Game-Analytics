```python
import json
import requests
import pandas as pd
```

```python
# Getting Complexes data

url = "https://api.sportradar.com/tennis/trial/v3/en/complexes.json?api_key=0eZfjbpWclEnUeESpwHbUce8a98A7IoGghY9w9GL"

answers = requests.get(url)

data = answers.json()

data
```

```python
# Complexes List
complexes_list = []
for comp in data["complexes"]:
    complexes_list.append(
        {
            "complex_id": comp["id"],
            "complex_name": comp["name"]
        }
    )
complexes_list
complexes_df = pd.DataFrame(complexes_list)
complexes_df.to_csv("complexes_df.csv",index = False)
```

```python
data
```

```python
# Venue List
venues_list = []
for complex in data.get("complexes",[]):
    for venue in complex.get("venues",[]):
        venues_list.append(
            {
                "venue_id" : venue.get("id"),
                "venue_name" : venue.get("name"),
                "city_name" : venue.get("city_name"),
                "country_name": venue.get("country_name"),
                "country_code": venue.get("country_code"),
                "timezone" : venue.get("timezone"),
                "complex_id": complex.get("id")
            }
        )
venues_list
venue_df = pd.DataFrame(venues_list)
venue_df.to_csv("venue_df.csv",index = False)
```
