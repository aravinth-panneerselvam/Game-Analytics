```python
import pandas as pd
import json
import requests
```

```python
url = "https://api.sportradar.com/tennis/trial/v3/en/competitions.json?api_key=0eZfjbpWclEnUeESpwHbUce8a98A7IoGghY9w9GL"
response = requests.get(url)

source_data = response.json()
source_data
```

```python
data = source_data["competitions"]
data
```

```python
# Getting competition data
competition_list= []
for comp in data:
    if "parent_id" not in comp:
        competition_list.append(
        {
            "id": comp["id"],
            "name": comp["name"],
            "type": comp["type"],
            "gender": comp["gender"]
        }   
        )
    else:
        competition_list.append(
        {
            "id": comp["id"],
            "name": comp["name"],
            "parent_id": comp["parent_id"],
            "type": comp["type"],
            "gender": comp["gender"]
        }   
        )
competition_list
```

```python
# Getting category data
category_list = []
for comp in data:
    category_list.append(
        {
            "category_id": comp["category"]["id"],
            "category_name": comp["category"]["name"]
        }
    )
category_list
```

```python
# GENERATING CATEGORY TABLE
category_df = pd.DataFrame(category_list)
category_df
```

```python
category_ref = category_df.drop_duplicates(subset="category_name")
category_ref
```

```python
category_ref.to_csv("category_ref.csv",index = False)
category_df.to_csv("category_df.csv",index = False)
```

```python
# GENERATING COMPETITION TABLE
competition_df = pd.DataFrame(competition_list)
competition_df.drop_duplicates(keep=False)
competition_df["category_id"] = category_df["category_id"]
competition_df.to_csv("competition_df.csv",index = False)
```
